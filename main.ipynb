### Elif
Hagamos un programa en el que preguntamos la calificacion del 0 al 100 y queremos saber su equivalente en el sistema de evaluacion eeuu. Para siplicidad, basemonos en esto:
- A+=100
- A=90-99
- B=80-89
- C=70-79
- D=60-69
- F=cualquier menor a 60



```python
cal = 78
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else: # notemos como al ya no haber una condicion el else cubre todo lo demas
  print("F")
```

## Operadores lógicos con condicionales

```python
# and, or, not
# Quieres un Sandwich\n",
pan = True
queso = True
mayonesa = True
jamón = False
print(pan and queso and mayonesa and jamón)
if not (pan and queso and mayonesa and jamón):
  print("No te puedes hacer un sandwich")
else:
  print("Puedes hacer un sandwich")
```

```python
cal = -5
if cal >= 0 and cal <=100: # Se secciona para ``oner un limite de no poder poner numero negativos
  if cal == 100:
    print("A+")
  elif cal >= 90:
    print("A")
  elif cal >= 80:
    print("B")
  elif cal >= 70:
    print("C")
  elif cal >= 60:
    print("D")
  else: # notemos como al ya no haber una condicion el else cubre todo lo demas
    print("F")
else:
  print("Error: la calificación debe estar entre 0 y 100")
```

```python
# El módulo (%)
# El módulo devuelve el residuo de una división

num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El numero {num} es múltiplo de 3 0 de 5")

# saber si el numero es par o impar usando not
num = 6
if not num % 2 == 0:
  print(f"El numero {num} es impar")
else:
  print(f"El numero {num} es par")
```

## CICLO FOR
## los llamados ciclos for, nos ayuda a realizar acciones durante una cantidad de veces

```python
# Imprimir numeros de 0 a 4
for i in range(3): # Toma 5 como el numero final pero inicia SIEMPRE en 0
  print(i)
```

```python
for num in range(2,5):
  print (num)
```

```python
# Tambien existe el range (inicio, fin, paso)
for i in range (2, 21, 3):
  print (i)
```

```python
cadena = "Python"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

## While


```python
suma = 0
números = int(input("Ingresa número para sumar o 0 para salir"))

while números != 0: # ! = es equivalente a "distinto de..."
  suma += números
  números = int(input("Ingresa número para sumar o 0 para salir"))

print(f"La suma es {suma}")
```

```python
# Usar el while como for
numero = 0
while numero <= 5:
  print(numero)
  numero += 1
```
